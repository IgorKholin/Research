#### Обработка данных

Здесь формат данных немного отличается, в большинстве случаев встречается время, в том числе сотые секунды, для этого пишем специальный формат:   
  
'%Y-%m-%d %H:%M:%S.%f  
Чтобы не потерять данные, возможно, имеет необходимость распределение по часам платежей/кредитов

```python
# приведение типов данных
plan['plan_at'] = pd.to_datetime(plan['plan_at'], format='%Y-%m-%d')
orders['created_at'] = pd.to_datetime(orders['created_at'], format='%Y-%m-%d %H:%M:%S.%f')
orders['put_at'] = pd.to_datetime(orders['put_at'], format='%Y-%m-%d %H:%M:%S.%f')
orders['closed_at'] = pd.to_datetime(orders['closed_at'], format='%Y-%m-%d %H:%M:%S.%f')
payments['paid_at'] = pd.to_datetime(payments['paid_at'], format='%Y-%m-%d %H:%M:%S.%f')
payments['order_id'] = payments['order_id'].astype('int')
plan['order_id'] =plan['order_id'].astype('int')
orders['order_id'] =orders['order_id'].astype('int')
```

```python
# проверка корректности записи данных. актуальная дата отчёта 2022-12-09
orders[orders['closed_at'] >= '2022-12-09']
```

```python
# актуализирруем данные, удаляя строки с будущими платежами 
plan = plan[plan['plan_at'] <= '2022-12-09']
len(plan)
```

#### Доработка данных, создание фичей

```python
# закрыт ли кредит 1 - если закрыт, 0 - нет
orders['closed'] = orders['closed_at'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
# возьмём заказ, который нам расскажет целую историю
for table in tables:
    display(table[table['order_id']== 400001732])
```

```python
# доработка фичи - внутрь каждого кредита добавляем айди операции - по плановым и по платёжным, чтобы в дальнейшем соеденить 
plan['inner_id'] = plan.groupby('order_id')['plan_at'].rank(method='dense') # используем окно с ранжированием без пропусков
payments['inner_id'] = payments.groupby('order_id')['paid_at'].rank(method='dense')

# соединяю исходя из айди кредита и из айди внутренней операции(1 - первая операцияб 2 - вторая и так далее)
pay_plan = plan.merge(payments, how='left', on=['order_id', 'inner_id'])

# считаем суммы каммулятинвым итогом - сколько было оплачено
pay_plan['total_paid'] = pay_plan.groupby('order_id')['paid_sum'].cumsum()
# добавляем признак закрытия кредита и сумму первоначального займа
pay_plan = pay_plan.merge(orders[['order_id','issued_sum','closed']], on='order_id', how='left')
pay_plan
```

Платёж по кредиту просрочен в двух случаях:  
1) сумма платежа поступила не в полном объёме  
2) Платёж по ступил позднее планового срока  
  
При этом стоить учитывать, то, что платёж мог поступить не полностью, но раньше планового срока, в таком случае это не просрочка  
Выразим данные условия в виде условия в np.where

```python
# признак просрочки
pay_plan['overdue'] = np.where(((pay_plan['total_paid'] - pay_plan['plan_sum_total'] < 0) &  # ситуация - обещанный платёж поступил не полностью
                            ((pay_plan['paid_at'] - pay_plan['plan_at']).dt.days >= 0)) | # ситуация - платёж не поступил в срок
                            ((pay_plan['paid_at'] - pay_plan['plan_at']).dt.days > 0),  
                             1, 0)
pay_plan[pay_plan['overdue'] == 1].head(5)
```

```python
# здесь составляю общую аналитическую сводку по доле просроченных платежей внутри кредита, далее рассматриваю возможные корреляции,
# а также распределения данных
analytics_data = pay_plan.groupby('order_id').agg(
    pays_acts = ('overdue', 'count'), 
    overdue=('overdue', 'mean'),  # процент просроченных платежей внутри кредита
    closed=('closed', 'max'), 
    paid_sum=('paid_sum', 'max'), 
    issued_sum=('issued_sum', 'max'))
```

```python
sns.histplot(analytics_data.overdue, kde=True, bins=25)
plt.title('Визуализация распределения доли просрочек внутри займа')
plt.ylabel('Кол-во займов')
plt.xlabel('Доля просроченных платежей')
```

```python
overdue_only = analytics_data[analytics_data.overdue != 0] # берём займы где были просрочки
sns.histplot(overdue_only.overdue, kde=True, bins=25)
plt.title('Визуализация распределения доли просрочек внутри займа(только займы с просрочками)')
plt.ylabel('Кол-во займов')
plt.xlabel('Доля просроченных платежей')
```

```python
# смотрим что влияет на просрочку платежа
sns.pairplot(analytics_data[['pays_acts','overdue','issued_sum', 'paid_sum', 'closed']], hue='closed')
```

Явных корреляций не наблюдается, попробуем посмотреть отдельные элменты, например, влияет ли сумма кредита на долю просроченых платежей

```python
sns.jointplot(analytics_data, x=analytics_data.issued_sum, y=analytics_data.overdue, hue='closed')
plt.title('Взаимосвязь % просрочененных платежей и суммы займа')
plt.ylabel('Доля просроченных платежей в займе')
plt.xlabel('Сумма займа')
```

```python
# делаем разбивки и считаем корреляции
import scipy.stats as stats

closed0 = analytics_data[analytics_data['closed']==0] # незакрытые кредиты
closed1 = analytics_data[analytics_data['closed']==1]
# для корреляции
r0, p_value = stats.pearsonr(closed0['issued_sum'], closed0['overdue'])
r1, p_value = stats.pearsonr(closed1['issued_sum'], closed1['overdue'])

g0 = sns.jointplot(closed0, y=closed0.issued_sum, x=closed0.overdue, kind='reg')
g0.ax_joint.annotate(f'r = {r0:.2f}', xy=(0.7, 0.9), xycoords='axes fraction', fontsize=12)
plt.xlabel('Доля просроченных платежей в займе')
plt.ylabel('Сумма займа')
# незакрытые кредиты
```

```python
g1 = sns.jointplot(closed1, y=closed1.issued_sum, x=closed1.overdue, kind='reg')
g1.ax_joint.annotate(f'r = {r1:.2f}', xy=(0.7, 0.9), xycoords='axes fraction', fontsize=12)
plt.xlabel('Доля просроченных платежей в займе')
plt.ylabel('Сумма займа')
# закрытые кредиты
```

Итак, корреляция есть, но она не выражена явным образом, и коллеблится от -7% до 4%. Следовательно, размер кредита, практически не влияет на долю просроченных платежей.  
Однако, стоит обратить внимание, что суммы кредитов небольшие, 50% из которых колеблется в диапазоне от ~4к до ~11к(ящик с усами займов), поэтому имеет место быть фактор случайности, нежели последовательности/системности
